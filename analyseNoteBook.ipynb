Récupérer les 5 plus grand vendeur

```python
globalSalesByPublisher = df.groupby(['publisher'])['global_sales'].count()
globalSalesByPublisher.nlargest(10)

```

Récupérer le nombre de jeu vendu par platforme

```python
globalSalesByPlatform = df.groupby(['platform'])['global_sales'].count()
globalSalesByPlatform.nlargest(5)
```

```python
data['% jp_sales'] = data['jp_sales'] / data['global_sales']
data['% eu_sales'] = data['eu_sales'] / data['global_sales']
data['% na_sales'] = data['na_sales'] / data['global_sales']


data.sort_values(by='global_sales', ascending=False).head(10)
```

C'est souvent en Amérique du Nord qu'il y a une plus grande part de jeu vendu 

Récuperer le nombre de jeu vendu par genre

```python
globalSalesByGenre = df.groupby(['genre'])['global_sales'].count()
globalSalesByGenre.nlargest(10)
```

```python
df_toGraph = data.groupby(by='genre', as_index=False).agg({'name': pd.Series.nunique})
df_toGraph = df_toGraph.sort_values(by=['name'], ascending=False)
df_toGraph.rename(columns = {'name':'video_games_count'}, inplace = True)
df_toGraph
```

```python
plt.figure(figsize=(12,10))

sns.barplot(x = 'video_games_count', y = 'genre', data = df_toGraph )
plt.title("Video Games Counts by Genre", fontsize=20)
plt.xlabel("Number of Video Games", fontsize=18)
plt.ylabel("Genre", fontsize=18)
plt.yticks(fontsize=14)
plt.show()
```

Le nombre de ventes globales des jeux selon les genres dépend énormement du nombre de jeux vendus par genre (le classement est quasiment le même)

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(data.corr(), vmin=-1, vmax=1, annot=True)
```
