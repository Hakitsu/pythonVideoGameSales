```python
df_filter = df.dropna()
df = df_filter[df_filter.publisher != 'Unknown']
df
```

##### recupérer les 50 éditeurs qui ont publié dans liste

```python
df_publisher = (df.groupby(["publisher"]).count().sort_values("rank", ascending=False)['rank']).nlargest(50)
df_publisher
```

```python
df_publisher_top = df["publisher"].value_counts().head(50).index
df_publisher_top
```

```python
def replace_publiser(x):
    if x["publisher"] in df_publisher_top :
        return x["publisher"]
    else : 
        return "Other"
df["publisher"] = df.apply(replace_publiser, axis = 1)
```

```python
df
```

```python
enc = OneHotEncoder(drop ='if_binary')
data_enc = enc.fit_transform(df["genre"].values.reshape(-1, 1)).toarray()
column_enc = enc.get_feature_names_out(["genre"])
one_hot_genre = pd.DataFrame(data_enc, columns =column_enc, index = df.index )
```

```python
enc = OneHotEncoder(drop ='if_binary')
data_enc = enc.fit_transform(df["publisher"].values.reshape(-1, 1)).toarray()
column_enc = enc.get_feature_names_out(["publisher"])
one_hot_publisher = pd.DataFrame(data_enc, columns =column_enc, index = df.index )
```

```python
enc = OneHotEncoder(drop ='if_binary')
data_enc = enc.fit_transform(df["platform"].values.reshape(-1, 1)).toarray()
column_enc = enc.get_feature_names_out(["platform"])
one_hot_platform = pd.DataFrame(data_enc, columns =column_enc, index = df.index )
```

```python
#df_filter_not_unknown[column_enc] = one_hot_genre

df_new_array = pd.concat([df, one_hot_platform, one_hot_genre, one_hot_publisher], axis = 1)
df_new_array
```

```python
(df.groupby(["publisher"]).count().sort_values("rank", ascending=False)['rank']).nlargest(50)
```

```python
df_new_array
```

```python
y_df = df_new_array["eu_sales"]
y_df
```

```python
x_df = df_new_array.drop(['rank','name','platform','genre','publisher'], axis = 1)
x_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.3, random_state=123)
```

```python
X_train
```

```python
y_test
```

```python
model = XGBRegressor()
model.fit(X_train, y_train)
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X_test, y_test, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (scores.mean(), scores.std()) )
```

```python
model.predict(X_train.head(1))
```
